### First simple example

```python
from sknn.mlp import Regressor, Layer
import numpy as np
```

```python
nn = Regressor(
    layers=[
        Layer("Rectifier", units=100),
        Layer("Linear")],
    learning_rate=0.02,
    n_iter=10)
```

```python
input = np.random.uniform(-0.5, 0.5, (100, 2))
# target is the sum of pairs of inputs
target = (input[:, 0] + input[:, 1]).reshape(100, 1)
```

```python
nn.fit(input, target)
```
