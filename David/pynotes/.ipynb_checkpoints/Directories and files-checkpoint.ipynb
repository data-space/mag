```python
import os
```

```python
path = "/Users/David/Dropbox/Data/MillionSongSubset/data/"
files = os.scandir(path)
def get_filenames(path):
    for entry in os.scandir(path):
        if entry.is_dir():
            print(get_filenames(path+entry.name+"/"))
```

```python
get_filenames(path)
```
