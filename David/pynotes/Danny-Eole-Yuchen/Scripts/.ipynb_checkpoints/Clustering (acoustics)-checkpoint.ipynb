```python
mss_df = pd.read_pickle(save_load_path+'/mss_df.pkl')
mss_df.head()

list(mss_df.columns.values)
```

```python
mss_df_d=mss_df.drop(['artist_id','artist_latitude','artist_location','artist_longitude','artist_name','song_id',
                      'track','track_id','title','release'], axis=1)
mss_df_d=mss_df_d.dropna()
```

```python
from scipy.cluster.vq import kmeans,vq,kmeans2
from sklearn.externals import joblib
from sklearn.metrics import *
from numpy import random
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist, pdist
from matplotlib import pyplot as plt
from sklearn.metrics import silhouette_score
%matplotlib inline
```

```python
k_range=range(1,20)
k_means_var=[KMeans(n_clusters=k).fit(mss_df_d) for k in k_range]
centroids=[X.cluster_centers_ for X in k_means_var]
```

```python
D_k = [cdist(mss_df_d, cent, 'euclidean') for cent in centroids]
cIdx = [np.argmin(D,axis=1) for D in D_k]
dist = [np.min(D,axis=1) for D in D_k]

#average within cluster sum of squares
avgWithinSS = [sum(d)/mss_df_d.shape[0] for d in dist]

#within cluster sum of squares
wcss = [sum(d**2) for d in dist]

#total sum of squares
tss = sum(pdist(mss_df_d)**2)/mss_df_d.shape[0]

#between cluster sum of squares
bss = tss-wcss

kIdx = 10-1


```

```python
# elbow curve
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(K, avgWithinSS, 'b*-')
ax.plot(K[kIdx], avgWithinSS[kIdx], marker='o', markersize=12, 
markeredgewidth=2, markeredgecolor='r', markerfacecolor='None')
plt.grid(True)
plt.xlabel('Number of clusters')
plt.ylabel('Average within-cluster sum of squares')
plt.title('Elbow for KMeans clustering')
```

```python
# number of clusters vs percentage of variance explained plot
fig=plt.figure()
ax = fig.add_subplot(111)
ax.plot(k_range, bss/tss*100, 'b*-')
ax.set_ylim((0,100))
plt.grid(True)
plt.xlabel('n_clusters')
plt.ylabel('Percentage of variance explained')
plt.title('Variance Explained vs k')
```

```python
# silouette coefficient plot

s = []
for n_clusters in range(2,20):
    kmeans = KMeans(n_clusters=n_clusters)
    kmeans.fit(mss_df_d)

    labels = kmeans.labels_
    centroids = kmeans.cluster_centers_

    s.append(silhouette_score(mss_df_d, labels, metric='euclidean'))

plt.plot(s)
plt.ylabel("Silouette")
plt.xlabel("k")
plt.title("Silouette for K-means cell's behaviour")
```

```python
random.seed(10)
#choose the number of clusters, 10
num_clusters = 10

km = KMeans(n_clusters=num_clusters)

%time km.fit(mss_df_d)

#dump the cluster into pickle file and load
joblib.dump(km,  'cluster.pkl')

km = joblib.load('cluster.pkl')

clusters = km.labels_
```
