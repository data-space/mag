## Import user-defined stop words

```python
import csv
from itertools import chain


list_csv = []

print("Loading stop words...")
with open(save_load_path+'/StopWords/SMART_stop_words.csv', 'r') as f:
    reader = csv.reader(f)
    list_csv = list(reader)
    
smart_stop_words = list(chain.from_iterable(list_csv))

print("Number of new stop words: {len_stop} \n".format(len_stop = len(smart_stop_words)))

print('Sample:')
for i in range(20,40):
    print(smart_stop_words[i])
```

## Remove custom stop-words from lyrics dataframe

```python
print('Removing stop words...')
print('Number of terms before removing stop words: {df_width}'.format(df_width = lyrics_raw_df.shape[1]))

for stopword in smart_stop_words:
    if stopword in lyrics_raw_df.columns.values.tolist():
        lyrics_raw_df = lyrics_raw_df.drop(stopword, 1)

print('Number of terms after removing stop words: {df_width}'.format(df_width = lyrics_raw_df.shape[1]))
```

## List all track_id with non zero amount of foreign words

```python
existing_terms = lyrics_raw_df.columns.values.tolist()

foreign_songs = []
associated_words = []

row_iterator = lyrics_raw_df.iterrows()

for i, row in row_iterator:
    foreign = False
    foreign_word = None
    
    for foreign_word in foreign_words:
        if foreign_word in existing_terms:
            if(row[foreign_word] > 0):
                foreign = True
                associated_words.append(foreign_word)
                break
    if foreign:
        foreign_songs.append(i)
    
print("Number of foreign songs detected: {num_foreign_songs}".format(num_foreign_songs = len(foreign_songs)))
```
