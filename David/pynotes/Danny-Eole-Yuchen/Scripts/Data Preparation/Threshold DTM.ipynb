```python
import pickle
import pandas as pd
import numpy as np
```

```python
def termThreshold(dtm_df, threshold):
    
    # Build array for the count of non-0 term-scores for each term
    sum_vec = dtm_df[ dtm_df > 0 ].count(axis=0)
    
    # Initialize list containing position of term columns to drop from dtm
    term_drop = []

    c = 0
    for column in dtm_df:
        if (len(column) < 3) or (sum_vec[c] < threshold) : 
            term_drop.append(column)
        c +=1
    
    dtm_df.drop(dtm_df.columns[term_drop], axis=1, inplace=True)
    return(dtm_df)
```

```python
# save cleaned dataframe as pickle file
mss_english_df = pd.read_pickle(save_load_path+'/lyrics_english_df.pkl')
```

```python
mss_english_thre2_df = termThreshold(mss_english_df, 2)
mss_english_thre3_df = termThreshold(mss_english_df, 3)
mss_english_thre4_df = termThreshold(mss_english_df, 4)
```
