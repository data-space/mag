```python
import pandas as pd
```

```python
import sqlite3

# Define path
## lyrics_path = '/home/eolus/Documents/MA755_data/LyricsData'
lyrics_path = '/Users/YuchenZhou/Documents/FILES/STUDY/755Assignments'

# Put `lyrics` table from `mxm_dataset.db` into lyrics_df pandas df
con = sqlite3.connect(lyrics_path +'/mxm_dataset.db')
lyrics_df = pd.read_sql_query("SELECT * from lyrics", con)
con.close()

lyrics_df.shape
```

```python
lyrics_df.head()
```

```python
mss_lyrics_df = lyrics_df[['track_id', 'word', 'count']]
```

```python
# Next, load the stopwords from the `nltk` package.
import nltk 
stopword_set = set(nltk.corpus.stopwords.words('english'))
is_not_stopword = [word not in stopword_set for word in mss_lyrics_df.word.ravel()]
```

```python
# Create dataframe `lyrics_sub_df` by removing the stopwords.
lyrics_clean_data = mss_lyrics_df[is_not_stopword]
```

```python
# We have now 4884 unique words for all the lyrics, which doesn't seem like much. 
len(pd.unique(lyrics_clean_data.word.ravel()))
```

```python
lyrics_clean_data.head()
lyrics_clean_data.shape
```

```python
# If we want more variables later then they should be merged into the dataframe.
# Now we _gather_ the rows into columns. __R__ users will recall the `tidyr` package.
mss_lyrics_pvt_df = lyrics_clean_data.pivot(index='track_id', 
                                        columns='word', 
                                        values='count')
```

```python
# Replace NaN values with 0
import numpy as np
NaN_locations = np.isnan(mss_lyrics_pvt_df)
mss_lyrics_pvt_df[NaN_locations] = 0

mss_lyrics_pvt_df.shape
```

```python
cutoff = int(round(len(mss_lyrics_pvt_df) / 10, 0))
cutoff
```

```python
mss_lyrics_1 = mss_lyrics_pvt_df.ix[range(cutoff)]
mss_lyrics_2 = mss_lyrics_pvt_df.ix[range(cutoff, cutoff*2)]
mss_lyrics_3 = mss_lyrics_pvt_df.ix[range(cutoff*2, cutoff*3)]
mss_lyrics_4 = mss_lyrics_pvt_df.ix[range(cutoff*3, cutoff*4)]
mss_lyrics_5 = mss_lyrics_pvt_df.ix[range(cutoff*4, cutoff*5)]
mss_lyrics_6 = mss_lyrics_pvt_df.ix[range(cutoff*5, cutoff*6)]
mss_lyrics_7 = mss_lyrics_pvt_df.ix[range(cutoff*6, cutoff*7)]
mss_lyrics_8 = mss_lyrics_pvt_df.ix[range(cutoff*7, cutoff*8)]
mss_lyrics_9 = mss_lyrics_pvt_df.ix[range(cutoff*8, cutoff*9)]
mss_lyrics_10 = mss_lyrics_pvt_df.ix[range(cutoff*9, mss_lyrics_pvt_df.shape[0])]
```

```python
mss_lyrics = mss_lyrics_1
mss_lyrics.append([mss_lyrics_2,mss_lyrics_3,mss_lyrics_4,mss_lyrics_5,mss_lyrics_6,mss_lyrics_7,mss_lyrics_8,mss_lyrics_9,
                   mss_lyrics_10],ignore_index=True)
```
